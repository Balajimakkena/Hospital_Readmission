```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\balaj\Desktop\Hospital_Readmission\Data\cleaned_data.csv")

print("✅ Data Loaded!")
print("Rows    :", df.shape[0])
print("Columns :", df.shape[1])
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='readmitted', data=df, 
              palette=['green', 'orange', 'red'])

plt.title('Patient Readmission Distribution', fontsize=15)
plt.xlabel('Readmission Status')
plt.ylabel('Number of Patients')

for p in plt.gca().patches:
    plt.gca().annotate(f'{int(p.get_height()):,}', 
                       (p.get_x() + p.get_width()/2, p.get_height()),
                       ha='center', va='bottom', fontsize=12)

plt.tight_layout()
plt.savefig(r"C:\Users\balaj\Desktop\Hospital_Readmission\readmission_distribution.png")
plt.show()
print("✅ Plot Saved!")
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='age', hue='readmitted', data=df,
              palette=['green', 'orange', 'red'])

plt.title('Age vs Readmission', fontsize=15)
plt.xlabel('Age Group')
plt.ylabel('Number of Patients')
plt.legend(title='Readmitted', loc='upper right')
plt.xticks(rotation=45)

plt.tight_layout()
plt.savefig(r"C:\Users\balaj\Desktop\Hospital_Readmission\age_vs_readmission.png")
plt.show()
print("✅ Age Plot Saved!")
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='gender', hue='readmitted', data=df,
              palette=['green', 'orange', 'red'])

plt.title('Gender vs Readmission', fontsize=15)
plt.xlabel('Gender')
plt.ylabel('Number of Patients')
plt.legend(title='Readmitted')

plt.tight_layout()
plt.savefig(r"C:\Users\balaj\Desktop\Hospital_Readmission\gender_vs_readmission.png")
plt.show()
print("✅ Gender Plot Saved!")
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='readmitted', y='time_in_hospital', data=df,
            palette=['green', 'orange', 'red'])

plt.title('Time in Hospital vs Readmission', fontsize=15)
plt.xlabel('Readmission Status')
plt.ylabel('Days in Hospital')

plt.tight_layout()
plt.savefig(r"C:\Users\balaj\Desktop\Hospital_Readmission\time_vs_readmission.png")
plt.show()
print("✅ Time Plot Saved!")
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='readmitted', y='num_medications', data=df,
            palette=['green', 'orange', 'red'])

plt.title('Number of Medications vs Readmission', fontsize=15)
plt.xlabel('Readmission Status')
plt.ylabel('Number of Medications')

plt.tight_layout()
plt.savefig(r"C:\Users\balaj\Desktop\Hospital_Readmission\medications_vs_readmission.png")
plt.show()
print("✅ Medications Plot Saved!")
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='readmitted', y='number_diagnoses', data=df,
            palette=['green', 'orange', 'red'])

plt.title('Number of Diagnoses vs Readmission', fontsize=15)
plt.xlabel('Readmission Status')
plt.ylabel('Number of Diagnoses')

plt.tight_layout()
plt.savefig(r"C:\Users\balaj\Desktop\Hospital_Readmission\diagnoses_vs_readmission.png")
plt.show()
print("✅ Diagnoses Plot Saved!")
```

```python
plt.figure(figsize=(12, 8))

numerical_cols = ['time_in_hospital', 'num_lab_procedures', 
                  'num_procedures', 'num_medications',
                  'number_outpatient', 'number_emergency',
                  'number_inpatient', 'number_diagnoses',
                  'readmitted_binary']

corr = df[numerical_cols].corr()

sns.heatmap(corr, annot=True, fmt='.2f', 
            cmap='RdYlGn', linewidths=0.5)

plt.title('Correlation Heatmap', fontsize=15)
plt.tight_layout()
plt.savefig(r"C:\Users\balaj\Desktop\Hospital_Readmission\correlation_heatmap.png")
plt.show()
print("✅ Heatmap Saved!")
```
